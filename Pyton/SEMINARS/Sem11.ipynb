```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
np.arange(0, 10)
```

```python
def func(a, b, c, x):
    return a*x**2 + b*x + c
```

```python
a, b, c = 5, 10, -30
x = np.arange(-10, 10, 0.01)
```

```python
def find_sqrt(a, b, c):
    d = b**2 - 4*a*c
    if d < 0:
        return ('Корней нет',)
    elif d == 0:
        x = -b / (2*a)
        return (round(x, 2),)
    else:
        x1 = (-b + d ** 0.5) / (2 * a)
        x2 = (-b - d ** 0.5) / (2 * a)
        return (round(x1, 2), round(x2, 2))
```

```python
sqrt = find_sqrt(a, b, c)
```

```python
min_y = min(func(a, b, c, x))
```

```python
min_x = find_sqrt(a, b, c - min_y)
min_x = min_x[0]
```

```python
pos_down = np.arange(-10, sqrt[1], 0.01)
neg_down = np.arange(sqrt[1], min_x, 0.01)

neg_up = np.arange(min_x, sqrt[0], 0.01)
pos_up = np.arange(sqrt[0], 10, 0.01)

#корни
plt.plot(sqrt[0], 0, 'go', label = f'Корени ({sqrt[0]}, {sqrt[1]})')
plt.plot(sqrt[1], 0, 'go')
#вершина
plt.plot(min_x, min_y, 'rx', label = f'Вершина ({min_x}, {min_y})')


# pos_down
plt.rcParams['lines.linestyle'] = '-'
plt.plot(pos_down, func(a, b, c, pos_down), 'b', label = 'Убывает больше ноля')

#neg_down
plt.rcParams['lines.linestyle'] = '-.'
plt.plot(neg_down, func(a, b, c, neg_down), 'b', label = 'Убывает меньше ноля')

#neg_up
plt.rcParams['lines.linestyle'] = '-.'
plt.plot(neg_up, func(a, b, c, neg_up), 'y', label = 'Возрастает меньше ноля')

#pos_up
plt.rcParams['lines.linestyle'] = '-'
plt.plot(pos_up, func(a, b, c, pos_up), 'y', label = 'Возрастает больше ноля')


plt.legend()
plt.grid()
```
